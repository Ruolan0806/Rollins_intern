```python
# transform the categorical columns into something we could use in the algorithmns
le=preprocessing.LabelEncoder()
df['region']=le.fit_transform(list(df['Region']))
len(df['region'].unique())
le2=preprocessing.LabelEncoder()
df['industry']=le2.fit_transform(list(df['Industry']))
len(df['Industry'].unique())
le3=preprocessing.LabelEncoder()
df['program']=le3.fit_transform(list(df['ProgramName']))
len(df['ProgramName'].unique())
```

```python
le4=preprocessing.LabelEncoder()
df['product']=le4.fit_transform(list(df['Product_Materials_Applied']))
len(df['product'].unique())
```

```python
le5=preprocessing.LabelEncoder()
df['bait']=le5.fit_transform(list(df['Bait_Materials_Applied']))
le6=preprocessing.LabelEncoder()
df['equipment']=le6.fit_transform(list(df['Equipment_Materials_Applied']))
```
